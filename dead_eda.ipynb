```python
def song_counter(frame):
    song_counter= Counter()
    for idx, row in frame.iterrows():
        for song in row:

            if not isinstance(song,float): # filter out the NaN's
                song_counter[song.strip()] += 1 #remove leading space
    return song_counter
```

```python
def counter_to_dataframe(song_ct):
    song_count_df = pd.DataFrame(song_ct,index=range(len(song_ct)))
    song_count_df.drop_duplicates(inplace=True)
    song_count_df=song_count_df.T
    song_count_df.reset_index(inplace=True,drop=False)
    song_count_df.sort_values(by=0,inplace=True,ascending=False)
    return song_count_df
```

```python
def graph_counters(song_count_df,set_name,num_songs_shown = 20,):
    plt.style.use('fivethirtyeight')
    fig,ax1 = plt.subplots(figsize=(12,7))
    sns.barplot(song_count_df['index'].head(num_songs_shown)\
                ,song_count_df[0].head(num_songs_shown),ax=ax1)
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=30,ha="right")
    if set_name =='Encore':
        ax1.set_yticks([int(a) for a in ax1.get_yticks()])
    ax1.set_xlabel("Songs")
    ax1.set_ylabel("Times Played")
    ax1.set_title(f"Most {num_songs_shown} Common Songs in {set_name} Played by D+C");
    
```

```python
set1_cols = [col for col in df if col.startswith('set1')]
set2_cols = [col for col in df if col.startswith('set2')]
encore_cols = [col for col in df if col.startswith('encore')]
col_filters = [set1_cols,set2_cols,encore_cols]
set_names = ['Set 1','Set 2','Encore']
for set_name,colfilter in zip(set_names,col_filters):
#     print(f"{set_name}")
    frame = df[colfilter]
#     print(frame)
    temp_counter = song_counter(frame)
    temp_counter_frame = counter_to_dataframe(temp_counter)
    graph_counters(temp_counter_frame,set_name,num_songs_shown=10)
```

```python
for idx, row in length_shows_df.iterrows():
    length_shows_df.loc[idx,'set2_cumulative'] = row['Set1_songs'] + row['Set2_songs']
    length_shows_df.loc[idx,'encore_cumulative'] = row['set2_cumulative'] + row['Encore_songs']
```

```python
length_shows_df_new= length_shows_df.drop(['Set2_songs','Encore_songs'],axis=1)
length_shows_df_stacked_cumulative = length_shows_df_new.stack().reset_index()
length_shows_df_stacked_cumulative.columns = ['show','set','count']
length_shows_df_stacked_cumulative
```

```python
def top_x_shows(x):
    show_lengths = length_shows_df_stacked_cumulative\
            [length_shows_df_stacked_cumulative['set']=='encore_cumulative']
    show_lengths = show_lengths.sort_values(by='count',ascending=False)
    show_lengths = show_lengths['show'].head(x)
    
    
    return show_lengths.values
top_x_shows(6)
```
