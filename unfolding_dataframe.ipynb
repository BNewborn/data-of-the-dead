```python
def set1(row):
    set_songs = list()
    if not pd.isna(row):

        row = row.split("Set 1:")
        row_split = re.split(">|,",row[1])

    
        for song in row_split:
            if song != ',':
                if song: 
                    song = re.sub("\d+$", "", song)
                    set_songs.append(song)

    else:
        set_songs.append("NO SETLIST")
    
    return set_songs
```

```python
set1 = df[2].apply(set1)
set1_df =pd.DataFrame()
for idx,row in set1.iteritems():
    temp_df = pd.DataFrame(row).T
    set1_df = pd.concat([set1_df,temp_df])
set1_df.reset_index(drop=True,inplace=True)
set1_df = set1_df.add_prefix("set1_song_")
set1_df
```

```python
def set2(row):
    set_songs = list()
    if not pd.isna(row):

        row = row.split("Set 2:")
        row_split = re.split(">|,",row[1])
#         row_split = re.split("(,)",row[1])
    
        for song in row_split:
            if song != ',':
                if song:
                    song = re.sub("\d+$", "", song)
                    set_songs.append(song)
    else:
        set_songs.append("NO SETLIST")
    
    return set_songs
```

```python
set2 = df[3].apply(set2)
set2_df =pd.DataFrame()
for idx,row in set2.iteritems():
    temp_df = pd.DataFrame(row).T
    set2_df = pd.concat([set2_df,temp_df])
set2_df.reset_index(drop=True,inplace=True)
set2_df = set2_df.add_prefix("set2_song_")
set2_df
```

```python
def encore(row):
    set_songs = list()
    if not pd.isna(row):
#         print(row)
        row_split = row.split(":")
        
        encore_split = re.split(">|,",row_split[1])
        
        for encore in encore_split:
            encore = re.sub("\d+$", "", encore)
            set_songs.append(encore)
    else:
        set_songs.append("NO ENCORES")
        
    return set_songs
```
